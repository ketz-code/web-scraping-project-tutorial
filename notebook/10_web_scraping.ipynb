```python
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Download all the HTML content of the URL
# Select the resource to download
url = "https://en.wikipedia.org/wiki/Gun_death_and_violence_in_the_United_States_by_state"

# Request to download the file from the Internet
response = requests.get(url)
response
#if response.status_code != 200:
    #raise Exception(f"Error accessing the page: {response.status_code}")

# Check the response
    #print("Status:", response.status_code)
```

```python
if response:
    soup = BeautifulSoup(response.text, 'html.parser')
    soup
soup
```

```python
table = soup.find('table')
data = []
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())
for row in table.find_all("tr"):
    cells = row.find_all("td")
    cells = [cell.text.strip() for cell in cells]
    data.append(cells)
data
```

```python
data_copy = data.copy()
del data_copy[0]
data_copy
```

```python
df = pd.DataFrame(data=data_copy,columns=headers)
df
```

```python
df.info()
```

### TAREA SOLICITADA POR EL PROFESOR EN CLASES EN SUSTITUCIÓN DE LA SOLICITADA EN EL MATERIAL 4GEEKS

```python
# Eliminar los valores entre corchetes
df['Gun  suicide  rate'] = df['Gun  suicide  rate'].astype(str).str.replace(r'\[.*?\]', '', regex=True)
df['Suicide  rate'] = df['Suicide  rate'].astype(str).str.replace(r'\[.*?\]', '', regex=True)
df['Gun  homicide  rate'] = df['Gun  homicide  rate'].astype(str).str.replace(r'\[.*?\]', '', regex=True)
df['Homicide  rate'] = df['Homicide  rate'].astype(str).str.replace(r'\[.*?\]', '', regex=True)
df['%  gun at  home'] = df['%  gun at  home'].astype(str).str.replace(r'\[.*?\]', '', regex=True)

# Elimina el signo de porcentaje, convierte a número y divide por 100
df['%  gun at  home'] = df['%  gun at  home'].astype(str).str.replace('%', '', regex=False)
df['%  gun at  home'] = pd.to_numeric(df['%  gun at  home'], errors='coerce') / 100

 # Convierte la columna a numérica, convirtiendo valores no numéricos a NaN
df['Gun  suicide  rate'] = pd.to_numeric(df['Gun  suicide  rate'], errors='coerce')
df['Suicide  rate'] = pd.to_numeric(df['Suicide  rate'], errors='coerce')
df['Gun  homicide  rate'] = pd.to_numeric(df['Gun  homicide  rate'], errors='coerce')
df['Homicide  rate'] = pd.to_numeric(df['Homicide  rate'], errors='coerce')

df
```
